```python
import geopandas as gpd
import rtree
import pygeos
import mapclassify
import geobr
```

- A variável "gdf"(geo data frame) vai receber os municipios referentes ao Brasil todo e selecionar o ano de 2020

```python
gdf = geobr.read_municipality(code_muni='all', year=2020)
gdf.head()
```

- A variável "bsb" (Brasília) está recebendo a coluna "name_muni" com o nome do municipio e exibindo

```python
bsb = gdf[gdf.name_muni=='Brasília']
bsb
```

* Para visualização de BSB e a utilização de cores e informações sobre o municipio.

```python
bsb.explore(
  color= "gray",
  tooltip = "name_muni",
  name = "Brasília",
  tiles = "CartoDB positron"
)
```

Para pegar a base de Escolas do Brasil

```python
gdf_base = geobr.read_schools(year=2020)
gdf_base.head()
```

"gdf_base_bsb" = recebe a partir do geopandas a junção espacial (spatial join), utilizando o método sjoin()


```python
gdf_base_bsb = gpd.sjoin(gdf_base, bsb)
gdf_base_bsb.head()
```

- Ler as colunas do geo data frame

```python
gdf_base_bsb.columns
```

* Aqui esta sendo criado um mapa centrado em Brasília (BSB)
e definindo as configurações de fizualização do mapa, como cores, legenda e tamanho...

```python
m = bsb.explore(
  color= "gray",
  tooltip = "name_muni",
  name = "São Paulo",
  tiles = "CartoDB positron"
)

gdf_base_bsb.explore(
  m=m,
  column ="government_level",
  scheme = "naturalbreaks",
  cmap = "Spectral",
  tooltip = False,
  name = "Escolas",
  style_kwds=dict(color="black", weight=1, opacity= 0.4), ##Argumentos de estilização do mapa
  popup =["name_school", "address", "government_level", "phone_number"],
  legend=True, #mostrar legenda
  legend_kwds=dict(caption= 'Legenda'), #não use barra de cores e edite o nome da legenda
  marker_type="circle_marker",
  marker_kwds=dict(radius=5),
  height=600,
  width=600,
)
m
```

* Salvando o como um arquivo HTML

```python
m.save(outfile="escolas_bsb.html")
```
